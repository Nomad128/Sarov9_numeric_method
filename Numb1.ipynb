```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
t_step = 0.05
# t_step = 0.9
# b =  4.99
const = 1.1
dtype = np.float64
```

```python
an_t = np.arange(start=0.3,stop=1.,step=t_step,dtype=dtype)
an_x = (24/(12*an_t**2 - 4*an_t**3 + an_t**4 + const))
b = an_x[0]
# b = 1000
```

```python
number = len(an_t)
neiv_x = np.zeros(number,dtype=dtype)
neiv_t = np.copy(an_t)
neiv_x[0] = b
for i in range(number-1):
    neiv_x[i+1] = neiv_x[i] - (neiv_t[i]-neiv_t[i]**2/2+neiv_t[i]**3/6)*t_step*neiv_x[i]**2
```

```python
ivn_x = np.zeros(number,dtype=dtype)
ivn_t = np.copy(an_t)
ivn_x[0] = b

for i in range(number-1):
    t_t = ivn_t[i+1] - (ivn_t[i+1]**2)/2 + (ivn_t[i+1]**3)/6
    ivn_x[i+1] = (-1 + np.sqrt(1+ 4*t_step*t_t*ivn_x[i]))/(2*t_step*t_t)
```

```python
plt.figure(figsize=(10,6))
plt.plot(an_t,an_x,label = "anal")
plt.plot(neiv_t,neiv_x, label="ivn")
plt.plot(ivn_t,ivn_x, label="neiv")
plt.legend()
plt.show()
```
