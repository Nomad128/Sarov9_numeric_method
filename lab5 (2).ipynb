```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return(np.sin(x))
n=20
h=0.5
d=h
x=[0]*(n)
y=[0]*(n)
F=[0]*(n-1)
for i in range (0,n):
    x[i]=h*i
    y[i]=f(x[i])
for i in range(1,n-1):
    F[i]=(y[i+1]-y[i-1])*3/d
F[0]=2*(y[1]-y[0])/d
F.append(2*(y[n-2]-y[n-3])/d)

plt.scatter(x,y)

B0=1
C0=1
C=1
B=4
A=1
Bn=1
An=1
alpha=[0]*(n)
betta=[0]*(n)
alpha[1]=-C0/B0
betta[1]=F[0]/B0
for i in range(2,n):
    alpha[i]=-C/(A*alpha[i-1]+B)
    betta[i]=(F[i-1]-A*betta[i-1])/(A*alpha[i-1]+B)
a=[0]*(n)
a[n-1]=((F[n-1]-An*betta[n-1])/(An*alpha[n-1]+Bn))
for i in range(n-2,-1,-1):
    a[i]=alpha[i+1]*a[i+1]+betta[i+1]

#for i in range (1,n-1):
#    print(F[i])
#    print(A*a[i-1]+B*a[i]+C*a[i+1])

b=[0]*(n)
b[0]=2*(3*y[1]-3*y[0]-2*a[0]*d-a[1]*d)/(d**2)
for i in range(1,n):
    b[i]=2*(2*a[i]*d-3*y[i]+3*y[i-1]+a[i-1]*d)/(d**2)
c=[0]*(n)
for i in range (0,n-1):
    c[i]=6*(a[i+1]*d+a[i]*d-2*y[i+1]+2*y[i])/(d**3)
# c[0]=0
# c[n-1]=0
xs=np.linspace(0,h,n)
y_g=y[0]+a[0]*xs+b[0]*(xs**2)/2
y_g1=y[1]+a[1]*xs+b[1]*(xs**2)/2+c[1]*(xs**3)/6
plt.plot(xs, y_g)
plt.plot(xs+h, y_g1)
plt.show()
y_s=[0]*(n-1)
for i in range(0,n-1):
    y_s[i]=y[i]+a[i]*xs+b[i]*(xs**2)/2+c[i]*(xs**3)/6
    plt.plot(xs+i*h, y_s[i])
print(x)
print(y)
print(F)
print(alpha)
print(betta)
print(a)
print(b)
print(c)
```

```python
%matplotlib
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return(np.sin(x))
n=40
h=0.01
d=[0]*(n)
x=[0]*(n+1)
y=[0]*(n+1)
F=[0]*(n-1)
for i in range (0,n+1):
    x[i]=h*(i**2)
    y[i]=f(x[i])
for i in range(0,n):
    d[i]=x[i+1]-x[i]
for i in range(1,n-1):
    F[i]=(y[i+1]-y[i-1])*3/d[i]
F[0]=2*(y[1]-y[0])/d[i]
F.append(2*(y[n-2]-y[n-3])/d[n-1])

plt.scatter(x,y)

B0=1
C0=1
C=1
B=4
A=1
Bn=1
An=1
alpha=[0]*(n)
betta=[0]*(n)
alpha[1]=-C0/B0
betta[1]=F[0]/B0
for i in range(2,n):
    alpha[i]=-C/(A*alpha[i-1]+B)
    betta[i]=(F[i-1]-A*betta[i-1])/(A*alpha[i-1]+B)
a=[0]*(n)
a[n-1]=((F[n-1]-An*betta[n-1])/(An*alpha[n-1]+Bn))
for i in range(n-2,-1,-1):
    a[i]=alpha[i+1]*a[i+1]+betta[i+1]

#for i in range (1,n-1):
#    print(F[i])
#    print(A*a[i-1]+B*a[i]+C*a[i+1])

b=[0]*(n)
b[0]=2*(3*y[1]-3*y[0]-2*a[0]*d[0]-a[1]*d[0])/(d[0]**2)
for i in range(1,n):
    b[i]=2*(2*a[i]*d[i]-3*y[i]+3*y[i-1]+a[i-1]*d[i])/(d[i]**2)
c=[0]*(n)
for i in range (0,n-1):
    c[i]=6*(a[i+1]*d[i]+a[i]*d[i]-2*y[i+1]+2*y[i])/(d[i]**3)
#c[0]=0
#c[n-1]=0

y_s=[0]*(n-1)
for i in range(0,n-1):
    xs=np.linspace(0,d[i],100)
    y_s[i]=y[i]+a[i]*xs+b[i]*(xs**2)/2+c[i]*(xs**3)/6
    plt.plot(xs+x[i], y_s[i])
print(x)
print(y)
print(F)
print(alpha)
print(betta)
print(a)
print(b)
print(c)
```
